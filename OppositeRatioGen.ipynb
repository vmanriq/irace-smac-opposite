```python
import json
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loadJsons(algorithm):
    filesData = []
    for i in range(1, 6):
        with open(f"./smacData/{algorithm}-{i}-runhistory.json") as file:
            data = json.load(file)
            filesData.append(data)
    return filesData

def getRatios(filesData):
    allRatios = []
    for data in filesData:
        descartadasCount = list(data["config_origins"].values()).count("Descartada")
        ratio = descartadasCount/len(data["config_origins"])
        allRatios.append(ratio)
    return allRatios
```

### SMAC

```python
gaData = loadJsons('ga')
gaRatio = getRatios(gaData)
gaRatio
```

```python
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(['1', '2', '3', '4', '5'], gaRatio, color ='red',
        width = 0.4)
 
plt.xlabel("Semillas")
plt.ylabel("Porcentaje de configuraciones opuestas evaluadas")
plt.title("Porcentaje de configuraciones opuestas evaluadas GA")
plt.ylim(0, 1)
plt.show()
```

```python
acotspData = loadJsons('acotsp')
acotspRatio = getRatios(acotspData)
acotspRatio
```

```python
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(['1', '2', '3', '4', '5'], acotspRatio, color ='red',
        width = 0.4)
 
plt.xlabel("Semillas")
plt.ylabel("Porcentaje de configuraciones opuestas evaluadas")
plt.title("Porcentaje de configuracion opouestas evaluadas ACOTSP")
plt.ylim(0, 1)
plt.show()
```
